Mini Project :
Student Marks Program

```python
# Function to calculate total marks
def calculate_total(marks):
    return sum(marks)

# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks)

# Lambda function to find highest mark
highest_mark = lambda marks: max(marks)

# Taking student marks as input
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)

# Calling functions
total = calculate_total(marks)
average = calculate_average(marks)
highest = highest_mark(marks)

# Display results
print("\nResults:")
print("Marks:", marks)
print("Total Marks:", total)
print("Average Marks:", round(average, 2))
print("Highest Mark:", highest)
```

Mini Project :
AI Number Guessing Game Lite

```python
import random  # Import random module to generate secret number

# Generate a random number between 1 and 50
secret = random.randint(1, 50)

# Maximum attempts allowed
max_attempts = 7

# Counter to track number of attempts
attempt = 0

print("===== AI Number Guessing Game Lite =====\n")
print("Guess a number between 1 and 50")
print(f"You have {max_attempts} attempts\n")

# Loop until attempts are finished
while attempt < max_attempts:
    
    # Ask user for input
    guess_input = input("Enter your guess: ")
    
    # Validate input (must be a number)
    if not guess_input.isdigit():
        print("Invalid input! Please enter a number between 1 and 50.\n")
        continue
    
    # Convert input to integer
    guess = int(guess_input)
    
    # Validate range
    if guess < 1 or guess > 50:
        print("Please enter a number within 1 to 50.\n")
        continue
    
    # Increase attempt count
    attempt += 1
    
    # Check guess
    if guess > secret:
        print("Too high!")
    
    elif guess < secret:
        print("Too low!")
    
    else:
        print(f"\nYou win in {attempt} attempts!")
        print("AI training level: Beginner → Intermediate")
        break  # Exit loop if correct
    
    # Show remaining attempts
    remaining = max_attempts - attempt
    print(f"Remaining attempts: {remaining}\n")

# If user fails after all attempts
if guess != secret:
    print("\n===== Game Over =====")
    print(f"The secret number was: {secret}")
    print("Better luck next time!")
```
