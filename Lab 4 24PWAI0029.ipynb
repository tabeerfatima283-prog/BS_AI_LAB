```python
# =========================================
# AI Number Guessing Game Lite
# =========================================

import random

# Generate random number between 1 and 50
secret_number = random.randint(1, 50)

# Total attempts
max_attempts = 7

print(f"Welcome to AI Number Guessing Game Lite!")
print(f"You have {max_attempts} attempts to guess the number between 1 and 50.\n")

# Game loop
for attempt in range(1, max_attempts + 1):

    try:
        # User input
        guess = int(input(f"Attempt {attempt}: Enter your guess: "))

        # Validate range
        if guess < 1 or guess > 50:
            print(f"Please enter a number between 1 and 50.\n")
            continue

        # Correct guess
        if guess == secret_number:
            print(f"\nYou win in {attempt} attempts!")
            print(f"AI training level: Beginner → Intermediate")
            break

        # Too high
        elif guess > secret_number:
            print(f"Too high!")

        # Too low
        else:
            print(f"Too low!")

        # Remaining attempts
        remaining = max_attempts - attempt
        print(f"Remaining attempts: {remaining}\n")

    # Handle invalid input
    except ValueError:
        print(f"Invalid input! Please enter an integer.\n")

# If player loses
else:
    print(f"\nGame Over!")
    print(f"The correct number was: {secret_number}")
```
