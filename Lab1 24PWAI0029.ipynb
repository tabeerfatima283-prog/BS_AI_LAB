```python
# PART 1 – Variables

# Create variables
name = "Tabeer"
age = 19
city = "Mansehra"

# Print all values
print("Name:", name)
print("Age:", age)
print("City:", city)

# Create number variables
num1 = 10
num2 = 5

# Perform calculations
sum_result = num1 + num2
subtraction_result = num1 - num2
multiplication_result = num1 * num2

# Print results
print("\nNumber 1:", num1)
print("Number 2:", num2)
print("Sum:", sum_result)
print("Subtraction:", subtraction_result)
print("Multiplication:", multiplication_result)
```

```python
# PART 2 – Input and Output

# Take name and age input from user
name = input("Enter your name: ")
age = input("Enter your age: ")

# Display message
print(f"Hello {name}, you are {age} years old")

# Take two numbers from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform calculations
addition = num1 + num2
subtraction = num1 - num2

# Print results
print("Addition:", addition)
print("Subtraction:", subtraction)
```

```python
# Ask age and print after 10 years
age = int(input("Enter your age: "))
print("After 10 years, your age will be:", age + 10)

# Ask marks and print average
marks1 = float(input("Enter marks 1: "))
marks2 = float(input("Enter marks 2: "))
marks3 = float(input("Enter marks 3: "))

average_marks = (marks1 + marks2 + marks3) / 3
print("Average Marks:", average_marks)
```

```python
# Task 1 – Even or Odd
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("Number is Even")
else:
    print("Number is Odd")
```

```python
# Task 2 – Grade System
marks = float(input("Enter your marks: "))
if marks >= 80:
    print("Grade: A")
elif marks >= 60:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

```python
# Task 3 – Voting Eligibility
age = int(input("Enter your age for voting check: "))
if age >= 18:
    print("You can vote")
else:
    print("Not eligible")
```

```python
# Task 4 – Print Numbers (For Loop)
print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i)
```

```python
# Task 5 – Multiplication Table
num = int(input("Enter a number for multiplication table: "))
for i in range(1, 11):
    print(f"{num} × {i} = {num * i}")
```

```python
# Task 6 – Sum of First 10 Numbers
total = 0
for i in range(1, 11):
    total += i
print("Sum of first 10 numbers:", total)
```

```python
# Task 7 – Password Checker
correct_password = "toyota"
password = ""

while password != correct_password:
    password = input("Enter password: ")

print("Access Granted")
```

```python
# Task 8 – Function
def average(a, b, c):
    return (a + b + c) / 3

# Calling function
result = average(10, 20, 30)
print("Average using function:", result)
```
