```python
# =========================================
# AI Password Strength Checker Lite
# =========================================

# Total attempts
max_attempts = 5

print(f"Welcome to AI Password Strength Checker Lite!\n")

# Loop for attempts
for attempt in range(1, max_attempts + 1):

    # User input
    password = input(f"Attempt {attempt}: Enter your password: ")

    # Conditions
    score = 0

    # Condition 1: Length
    if len(password) >= 8:
        score += 1

    # Condition 2: Contains digit
    if any(char.isdigit() for char in password):
        score += 1

    # Condition 3: Contains uppercase letter
    if any(char.isupper() for char in password):
        score += 1

    # Feedback based on score
    if score == 0:
        strength = "Very Weak"
    elif score == 1:
        strength = "Weak"
    elif score == 2:
        strength = "Moderate"
    else:
        strength = "Strong"

    # Display feedback
    print(f"Password Strength: {strength}")

    # Winning condition
    if strength == "Strong":
        print(f"\nYou created a strong password!")
        print(f"AI Training Level: Beginner → Intermediate")
        break

    # Remaining attempts
    remaining = max_attempts - attempt
    print(f"Remaining Attempts: {remaining}\n")

# Losing condition
else:
    print(f"\nFinal Result: Password is not strong enough.")

```
