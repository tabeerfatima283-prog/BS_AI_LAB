```python
# This program asks the user for personal information
# and prints a formatted welcome message.

# Ask the user to enter their full name
full_name = input("Enter your full name: ")

# Ask the user to enter their favorite color
favorite_color = input("Enter your favorite color: ")

# Ask the user to enter their birth year
birth_year = int(input("Enter your birth year: "))

# Calculate age based on current year
current_year = 2026
age = current_year - birth_year

# Print welcome message using f-strings
print(f"\nWelcome, {full_name}!")

# Print user's favorite color
print(f"Your favorite color is {favorite_color} – perfect for calm AI thinking.")

# Print birth year and calculated age
print(f"You were born in {birth_year} → you are currently {age} years old ({current_year}).")
```

```python
# Mini calculator program with input validation

# Ask user for two numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Ask user for the operation
operation = input("Enter operation (+, -, *, /, **, %): ")

# Check if operation is valid
if operation not in ["+", "-", "*", "/", "**", "%"]:
    print("Invalid operator!")

# Addition
elif operation == "+":
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

# Subtraction
elif operation == "-":
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")

# Multiplication
elif operation == "*":
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")

# Division
elif operation == "/":
    if num2 == 0:
        print("Cannot divide by zero!")
    else:
        result = num1 / num2
        print(f"{num1} / {num2} = {result:.2f}")

# Power
elif operation == "**":
    result = num1 ** num2
    print(f"{num1} ** {num2} = {result:.2f}")

# Modulo
elif operation == "%":
    if num2 == 0:
        print("Cannot divide by zero!")
    else:
        result = num1 % num2
        print(f"{num1} % {num2} = {result}")
```

```python
# Ask user for starting and ending numbers
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

num = start

# Use while loop
while num <= end:

    # Skip numbers divisible by 7
    if num % 7 == 0:
        num += 1
        continue

    # Check conditions
    if num % 2 == 0 and num % 5 == 0:
        print(f"{num}: Even & Multiple of 5!!")

    elif num % 2 == 0:
        print(f"{num}: Even")

    elif num % 5 == 0:
        print(f"{num}: Multiple of 5!")

    else:
        print(num)

    num += 1
```

```python
# Ask user to enter password
password = input("Enter your password: ")

# Check password length
if len(password) < 6:
    print("Too short!")

else:
    # Check if password has a digit
    has_digit = any(char.isdigit() for char in password)

    # Check if password has uppercase letter
    has_upper = any(char.isupper() for char in password)

    # Nested conditions
    if not has_digit:
        print("Add at least one number")

    elif not has_upper:
        print("Add at least one capital letter")

    else:
        print("Strong password – good choice!")
```
